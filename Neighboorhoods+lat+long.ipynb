```python
df1=pd.read_csv('Data.csv')
```

Merging the two data frames

```python
df3.drop('Unnamed: 0', axis=1, inplace=True)
df3.drop('Postal Code', axis=1, inplace=True)
```

Downloading the given dataframe

```python
df3.to_csv('Neighborhood+lat+long.csv')
```

Making it according to the data frame given 

```python
Column_names = ["PostalCode", "Borough", "Neighborhood", "Latitude", "Longitude"]
df5 = pd.DataFrame(columns=Column_names)
list_postal_codes = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in list_postal_codes:
    df5 = df5.append(df3[df3["PostalCode"]==postcode], ignore_index=True)

df5
```

```python
df5.to_csv('shortened_lat_long.csv')
```
